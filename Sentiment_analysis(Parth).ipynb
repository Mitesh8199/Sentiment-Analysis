```python
import pandas as pd
tweets = pd.read_csv('train_tweets.csv')
test_tweets = pd.read_csv('test_tweets.csv')
tweets.head(10)
```

```python
X = tweets['tweet']
Y = tweets['label']
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
stop_words=stopwords.words('english')
stemmer=PorterStemmer()
```

```python
import re
cleaned_data=[]
for i in range(len(X)):
   tweet=(re.sub('[^a-zA-Z]',' ',X.iloc[i]))
   tweet=tweet.lower().split()
   tweet=[stemmer.stem(word) for word in tweet if (word not in stop_words)]
   tweet=' '.join(tweet)
   cleaned_data.append(tweet)
```

```python
print(cleaned_data)
```

```python
import nltk
nltk.download('wordnet')
from nltk.stem.wordnet import WordNetLemmatizer
def normalization(cleaned_data):
        lem = WordNetLemmatizer()
        normalized_tweet = []
        for word in cleaned_data:
            normalized_text = lem.lemmatize(word,'v')
            normalized_tweet.append(normalized_text)
        return normalized_tweet
    
print(normalization(cleaned_data))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3000,stop_words=['user'])
X_fin=cv.fit_transform(cleaned_data).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
```

```python
pipeline = Pipeline([
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

```python
msg_train, msg_test, label_train, label_test = train_test_split(X_fin,Y, test_size=0.2)
pipeline.fit(msg_train,label_train)
predictions = pipeline.predict(msg_test)
print("*NAIVE BAYES CLASSIFIER*")
print("-------Classification Report---------")
print(classification_report(predictions,label_test))
print("--------Confusion Matrix-------------")
print(confusion_matrix(predictions,label_test))
print('-----------Accuracy of the Model------------')
print(accuracy_score(predictions,label_test))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(msg_train)
sc.fit_transform(msg_test)
```

```python
def models(msg_train,label_train):

  #Logistic Regression
  from sklearn.linear_model import LogisticRegression
  log = LogisticRegression(random_state = 0)
  log.fit(msg_train,label_train)

  #Decision tree 
  from sklearn.tree import DecisionTreeClassifier
  tree = DecisionTreeClassifier(criterion = 'entropy',random_state=0)
  tree.fit(msg_train,label_train)

  #RandomForestClassifier
  from sklearn.ensemble import RandomForestClassifier
  forest = RandomForestClassifier(n_estimators = 10, criterion = "entropy", random_state = 0)
  forest.fit(msg_train,label_train)

  #Print the models accuracy on the training set
  print('[0]Logistic Regression Training Accuracy:',log.score(msg_train,label_train))
  print('[1]Decision Tree classifier Training Accuracy:',tree.score(msg_train,label_train))
  print('[2]Random forest classifier Training Accuracy:',forest.score(msg_train,label_train))

  return log,tree,forest
```

```python
model = models(msg_train,label_train)
```

```python
 #test model accuracy on test data on confusion matrix

from sklearn.metrics import confusion_matrix
for i in range(len(model)):
  print('Models',i)
  print("Confusion Matrix")
  cm = confusion_matrix(label_test,model[i].predict(msg_test))
  print(cm)
  print()
```
